```python
import pandas as pd
import os

# Path to cleaned CSVs
base_path = "processed_data"

csv_files = [
    "male_players_clean.csv",
    "male_players_legacy_clean.csv",
    "female_players_clean.csv",
    "female_players_legacy_clean.csv",
    "male_teams_clean.csv",
    "female_teams_clean.csv",
    "male_coaches_clean.csv",
    "female_coaches_clean.csv"
]

# Dictionary to hold array of columns per file
columns_dict = {}

for file in csv_files:
    file_path = os.path.join(base_path, file)

    print(f"\n==============================")
    print(f"ðŸ“„ FILE: {file}")
    print("==============================")

    try:
        df = pd.read_csv(file_path)

        # Convert columns to array (Python list)
        col_array = df.columns.tolist()

        # Store in dictionary
        columns_dict[file] = col_array

        # Print output nicely
        print(f"Total columns: {len(col_array)}")
        print("Columns array:")
        print(col_array)

    except Exception as e:
        print(f"âŒ Error reading file {file}: {e}")

# If you want to access specific arrays later:
# columns_dict["male_players_clean.csv"]
```

```python
import pandas as pd
import os

folder_path = r"null_removed"

files = [f for f in os.listdir(folder_path) if f.endswith(".csv")]

output = []

for file in files:
    df = pd.read_csv(os.path.join(folder_path, file), low_memory=False)
    nulls = df.isnull().sum()
    for col, n in nulls.items():
        output.append([file, col, n])

df_null = pd.DataFrame(output, columns=["dataset", "column", "null_count"])
df_null.to_csv("fifa_null_report.csv", index=False)

df_null
```

```python
import pandas as pd
import numpy as np

# ---------------------------------------
# UTILITY FUNCTIONS
# ---------------------------------------

def fix_date(df, col, fill=False):
    """Convert to datetime and optionally fill missing dates"""
    df[col] = pd.to_datetime(df[col], errors='coerce')
    if fill:
        df[col].fillna(pd.Timestamp("2012-01-01"), inplace=True)

def fix_int(df, cols):
    """Convert integer-like columns safely"""
    for col in cols:
        if col in df.columns:
            df[col] = df[col].fillna(0).astype('int64')

def fix_core_attributes(df):
    """Handle pace, shooting, pass/passing, dribbling, defending, physic safely"""
    
    # detect which 'pass' column is present
    possible_pass_cols = ["pass", "passing"]
    pass_col = [c for c in possible_pass_cols if c in df.columns]

    core_attrs = ["pace", "shooting", "dribbling", "defending", "physic"] + pass_col

    for col in core_attrs:
        if col in df.columns:
            df[col].fillna(0, inplace=True)

# ---------------------------------------
# 1. FEMALE COACHES
# ---------------------------------------

df = pd.read_csv("null_removed/fcoaches.csv", low_memory=False)
fix_date(df, "dob", fill=False)
df.to_csv("fcoaches.csv", index=False)


# ---------------------------------------
# 2. FEMALE PLAYERS
# ---------------------------------------

df = pd.read_csv("null_removed/fplayers.csv", low_memory=False)

fix_date(df, "fifa_update_date")
fix_date(df, "dob")
fix_date(df, "club_joined_date", fill=True)

df['value_eur'].fillna(0, inplace=True)
df['wage_eur'].fillna(0, inplace=True)
df['release_clause_eur'].fillna(0, inplace=True)

df['league_id'].fillna(0, inplace=True)
df['league_name'].fillna("No League", inplace=True)
df['league_level'].fillna(0, inplace=True)
df['club_team_id'].fillna(0, inplace=True)
df['club_name'].fillna("Free Agent", inplace=True)
df['club_position'].fillna("Unknown", inplace=True)
df['club_jersey_number'].fillna(0, inplace=True)
df['club_loaned_from'].fillna("None", inplace=True)
df['club_contract_valid_until_year'].fillna(0, inplace=True)

df['nation_team_id'].fillna(0, inplace=True)
df['nation_position'].fillna("Not Selected", inplace=True)
df['nation_jersey_number'].fillna(0, inplace=True)

df['player_tags'].fillna("None", inplace=True)
df['player_traits'].fillna("None", inplace=True)

fix_core_attributes(df)

df['mentality_composure'].fillna(df['mentality_composure'].median(), inplace=True)
df['goalkeeping_speed'].fillna(0, inplace=True)

fix_int(df, [
    'player_id','league_id','club_team_id','nation_team_id',
    'club_contract_valid_until_year','league_level',
    'club_jersey_number','nation_jersey_number'
])

df.to_csv("fplayers.csv", index=False)


# ---------------------------------------
# 3. FEMALE PLAYERS LEGACY
# ---------------------------------------

df = pd.read_csv("null_removed/fplayerslegacy.csv", low_memory=False)

fix_date(df, "fifa_update_date")
fix_date(df, "dob")
fix_date(df, "club_joined_date", fill=True)

df['value_eur'].fillna(0, inplace=True)
df['wage_eur'].fillna(0, inplace=True)
df['release_clause_eur'].fillna(0, inplace=True)

df['league_id'].fillna(0, inplace=True)
df['league_name'].fillna("No League", inplace=True)
df['league_level'].fillna(0, inplace=True)
df['club_team_id'].fillna(0, inplace=True)
df['club_name'].fillna("Free Agent", inplace=True)
df['club_position'].fillna("Unknown", inplace=True)
df['club_jersey_number'].fillna(0, inplace=True)
df['club_loaned_from'].fillna("None", inplace=True)
df['club_contract_valid_until_year'].fillna(0, inplace=True)

df['nation_team_id'].fillna(0, inplace=True)
df['nation_position'].fillna("Not Selected", inplace=True)
df['nation_jersey_number'].fillna(0, inplace=True)

df['player_tags'].fillna("None", inplace=True)
df['player_traits'].fillna("None", inplace=True)

fix_core_attributes(df)

df['mentality_composure'].fillna(df['mentality_composure'].median(), inplace=True)
df['goalkeeping_speed'].fillna(0, inplace=True)

fix_int(df, [
    'player_id','league_id','club_team_id','nation_team_id',
    'club_contract_valid_until_year','league_level',
    'club_jersey_number','nation_jersey_number'
])

df.to_csv("fplayerslegacy.csv", index=False)


# ---------------------------------------
# 4. FEMALE TEAMS
# ---------------------------------------

df = pd.read_csv("null_removed/fteams.csv", low_memory=False)
fix_date(df, "fifa_update_date")
df.to_csv("fteams.csv", index=False)


# ---------------------------------------
# 5. MALE COACHES
# ---------------------------------------

df = pd.read_csv("null_removed/mcoaches.csv", low_memory=False)
fix_date(df, "dob", fill=False)
df.to_csv("mcoaches.csv", index=False)


# ---------------------------------------
# 6. MALE PLAYERS
# ---------------------------------------

df = pd.read_csv("null_removed/mplayers.csv", low_memory=False)

fix_date(df, "fifa_update_date")
fix_date(df, "dob")
fix_date(df, "club_joined_date", fill=True)

df['value_eur'].fillna(0, inplace=True)
df['wage_eur'].fillna(0, inplace=True)
df['release_clause_eur'].fillna(0, inplace=True)

df['league_id'].fillna(0, inplace=True)
df['league_name'].fillna("No League", inplace=True)
df['league_level'].fillna(0, inplace=True)
df['club_team_id'].fillna(0, inplace=True)
df['club_name'].fillna("Free Agent", inplace=True)
df['club_position'].fillna("Unknown", inplace=True)
df['club_jersey_number'].fillna(0, inplace=True)
df['club_loaned_from'].fillna("None", inplace=True)
df['club_contract_valid_until_year'].fillna(0, inplace=True)

df['nation_team_id'].fillna(0, inplace=True)
df['nation_position'].fillna("Not Selected", inplace=True)
df['nation_jersey_number'].fillna(0, inplace=True)

df['player_tags'].fillna("None", inplace=True)
df['player_traits'].fillna("None", inplace=True)

fix_core_attributes(df)

df['mentality_composure'].fillna(df['mentality_composure'].median(), inplace=True)
df['goalkeeping_speed'].fillna(0, inplace=True)

fix_int(df, [
    'player_id','league_id','club_team_id','nation_team_id',
    'club_contract_valid_until_year','league_level',
    'club_jersey_number','nation_jersey_number'
])

df.to_csv("mplayers.csv", index=False)


# ---------------------------------------
# 7. MALE PLAYERS LEGACY
# ---------------------------------------

df = pd.read_csv("null_removed/mplayerslegacy.csv", low_memory=False)

fix_date(df, "fifa_update_date")
fix_date(df, "dob")
fix_date(df, "club_joined_date", fill=True)

df['value_eur'].fillna(0, inplace=True)
df['wage_eur'].fillna(0, inplace=True)
df['release_clause_eur'].fillna(0, inplace=True)

df['league_id'].fillna(0, inplace=True)
df['league_name'].fillna("No League", inplace=True)
df['league_level'].fillna(0, inplace=True)
df['club_team_id'].fillna(0, inplace=True)
df['club_name'].fillna("Free Agent", inplace=True)
df['club_position'].fillna("Unknown", inplace=True)
df['club_jersey_number'].fillna(0, inplace=True)
df['club_loaned_from'].fillna("None", inplace=True)
df['club_contract_valid_until_year'].fillna(0, inplace=True)

df['nation_team_id'].fillna(0, inplace=True)
df['nation_position'].fillna("Not Selected", inplace=True)
df['nation_jersey_number'].fillna(0, inplace=True)

df['player_tags'].fillna("None", inplace=True)
df['player_traits'].fillna("None", inplace=True)

fix_core_attributes(df)

df['mentality_composure'].fillna(df['mentality_composure'].median(), inplace=True)
df['goalkeeping_speed'].fillna(0, inplace=True)

fix_int(df, [
    'player_id','league_id','club_team_id','nation_team_id',
    'club_contract_valid_until_year','league_level',
    'club_jersey_number','nation_jersey_number'
])

df.to_csv("mplayerslegacy.csv", index=False)


# ---------------------------------------
# 8. MALE TEAMS
# ---------------------------------------

df = pd.read_csv("null_removed/mteams.csv", low_memory=False)
fix_date(df, "fifa_update_date")
df.to_csv("mteams.csv", index=False)

print("ðŸŽ‰ ALL 8 DATASETS CLEANED SUCCESSFULLY (MASTER SCRIPT v2)!")
```
