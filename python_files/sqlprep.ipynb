```python
import pandas as pd

fcoaches = pd.read_csv("null_removed/fcoaches.csv")
mcoaches = pd.read_csv("null_removed/mcoaches.csv")
```

```python
fcoaches['gender'] = 'F'
mcoaches['gender'] = 'M'
```

```python
coaches = pd.concat([fcoaches, mcoaches], ignore_index=True)
coaches.to_csv("coaches.csv", index=False)
```

```python
fplayers = pd.read_csv("null_removed/fplayers.csv")
mplayers = pd.read_csv("null_removed/mplayers.csv")
fplayerslegacy = pd.read_csv("null_removed/fplayerslegacy.csv")
mplayerslegacy = pd.read_csv("null_removed/mplayerslegacy.csv")

fplayers['gender'] = 'F'
mplayers['gender'] = 'M'
fplayers['is_legacy'] = 0
mplayers['is_legacy'] = 0

fplayerslegacy['gender'] = 'F'
mplayerslegacy['gender'] = 'M'
fplayerslegacy['is_legacy'] = 1
mplayerslegacy['is_legacy'] = 1

players = pd.concat([fplayers, mplayers, fplayerslegacy, mplayerslegacy], ignore_index=True)

players.to_csv("players.csv", index=False)
```

```python
fteams = pd.read_csv("null_removed/fteams.csv")
mteams = pd.read_csv("null_removed/mteams.csv")

fteams['gender']='F'
mteams['gender']='M'

teams = pd.concat([fteams,mteams], ignore_index=True)
teams.to_csv("teams.csv", index=False)
```
